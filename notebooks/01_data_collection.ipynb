#Soccer Player's Market Value Prediction
This project is to use data of soccer players to predict their market value 
##Data Source
[Football Data from Transfermarkt from Kaggle](https://www.kaggle.com/datasets/davidcariboo/player-scores?select=appearances.csv)
##Files Downloaded
* appearances.csv - 1706806 rows, 13 columns
* transfers.csv - 79646 rows, 10 columns
* players.csv - 32601 rows, 23 columns
* players_valuations.csv - 496606 rows, 5 columns
##Dataset description
The data includes players' information such as: age, club, market value, transfer fee, etc.
* 60,000+ games from many seasons on all major competitions
* 400+ clubs from those competitions
* 30,000+ players from those clubs
* 400,000+ player market valuations historical records
* 1,200,000+ player appearance records from all games

## Key Findings
- Target variable: market_value from player_valuations.csv (496,606 records)
- Datasets can be joined using player_id
- Appearance data needs aggregation (average per player)
